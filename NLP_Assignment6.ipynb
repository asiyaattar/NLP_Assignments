```python
# Advanced Sentiment Analysis using BERT

from transformers import pipeline

# Load sentiment analysis pipeline
sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

# Test sentences
sentences = [
    "I love this movie. It is amazing!",
    "This product is very bad and disappointing.",
    "The service was okay, not great.",
    "I am very happy with the results.",
    "Worst experience ever."
]

# Analyze sentiment
results = sentiment_analyzer(sentences)

# Display results
for text, result in zip(sentences, results):
    print("Text:", text)
    print("Sentiment:", result["label"])
    print("Confidence:", round(result["score"], 3))
    print("-" * 50)
```
